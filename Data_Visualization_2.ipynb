```python
df=pd.read_csv("/content/batter (1).csv")
df
```

```python
sample_df=df.head(50).sample(25)
sample_df
```

```python
plt.figure(figsize=(16,8))
plt.scatter(sample_df["avg"],sample_df["strike_rate"],s=sample_df["runs"])
plt.xlabel("average")
plt.ylabel("strike rate")
plt.axhline(130,color="red")
plt.axvline(35,color="red")
for i in range(sample_df.shape[0]):
  plt.text(sample_df["avg"].values[i],sample_df["strike_rate"].values[i],sample_df["batter"].values[i])
```

```python
fig,ax=plt.subplots(figsize=(10,6))
ax.scatter(df["avg"],df["strike_rate"])
ax.set_xlabel("average")
ax.set_ylabel("strike rate")
```

```python
fig,ax=plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,6))
ax[0].scatter(sample_df["runs"],sample_df["avg"],color="red",marker="D")

ax[0].set_xlabel("runs")
ax[0].set_ylabel("average")
ax[1].scatter(sample_df["runs"],sample_df["strike_rate"])
ax[1].set_xlabel("runs")
ax[1].set_ylabel("strike rate")
```
